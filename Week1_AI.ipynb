```python
img = Image.open("cry.jpg")
img_np = np.array(img)
```

```python
plt.figure(figsize=(4,4))
plt.imshow(img_np)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
top_left = img_np[0:100, 0:100]

plt.figure(figsize=(3,3))
plt.imshow(top_left)
plt.title("Top-left 100x100 Region")
plt.axis("off")
plt.show()
```

```python
R = img_np[:, :, 0]
G = img_np[:, :, 1]
B = img_np[:, :, 2]
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(R, cmap='Reds')
plt.title("Red Channel (R)")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(G, cmap='Greens')
plt.title("Green Channel (G)")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(B, cmap='Blues')
plt.title("Blue Channel (B)")
plt.axis("off")

plt.show()
```

```python
modified_img = img_np.copy()
modified_img[0:100, 0:100] = 210

plt.figure(figsize=(4,4))
plt.imshow(modified_img)
plt.title("Modified Image (Top-left = 210)")
plt.axis("off")
plt.show()
```

```python
img_color = Image.open("cry.jpg")
img_gray = img_color.convert("L")
gray_np = np.array(img_gray)

plt.figure(figsize=(4,4))
plt.imshow(gray_np, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

```python
# Get image dimensions
h, w = gray_np.shape

# Find center
center_h, center_w = h // 2, w // 2

# Extract 150x150 region
crop_size = 150
half = crop_size // 2

middle_section = gray_np[
    center_h - half:center_h + half,
    center_w - half:center_w + half
]

plt.figure(figsize=(3,3))
plt.imshow(middle_section, cmap="gray")
plt.title("Middle 150x150 Section")
plt.axis("off")
plt.show()
```

```python
binary_img = gray_np.copy()

binary_img[binary_img < 100] = 0
binary_img[binary_img >= 100] = 255

plt.figure(figsize=(4,4))
plt.imshow(binary_img, cmap="gray")
plt.title("Binary Image (Threshold = 100)")
plt.axis("off")
plt.show()
```

```python
rotated_pil = img_gray.rotate(-90, expand=True)

plt.imshow(rotated_pil, cmap="gray")
plt.title("Rotated (PIL)")
plt.axis("off")
plt.show()
```

```python
# Stack grayscale into 3 channels
rgb_from_gray = np.stack((gray_np,)*3, axis=-1)

plt.figure(figsize=(4,4))
plt.imshow(rgb_from_gray)
plt.title("Grayscale Converted to RGB")
plt.axis("off")
plt.show()
```

```python
# Compute mean
mean_vec = np.mean(img_np, axis=0)

# Center the data
X_centered = img_np - mean_vec

print("Data centered ✔")

cov_matrix = np.cov(X_centered, rowvar=False)
print("Covariance matrix shape:", cov_matrix.shape)
```
